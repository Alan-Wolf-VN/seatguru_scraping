```python
import bs4, requests, re
```

```python
import pandas as pd
```

```python
browse_airlines = requests.get('https://seatguru.com/browseairlines/browseairlines.php').text
```

```python
soup_browse_airlines = bs4.BeautifulSoup(browse_airlines,'lxml')
```

```python
airlines_links = soup_browse_airlines.find_all('a', href = re.compile("/airlines/")) #Dùng regular expression để tìm các href có /airlines/
```

```python
airlines_links
```

```python
airlines_list = []
for link in airlines_links:
    airlines_list.append(link['href'])
```

```python
airlines_list
```

```python
airlines_info_pages = []
for airlines in airlines_list:
    airlines_info_pages.append('https://seatguru.com'+airlines)
```

```python
airlines_info_pages
```

```python
seat_df = pd.DataFrame()
fleet_info_pages = []
for page in airlines_info_pages:
    airlines_soup = bs4.BeautifulSoup(requests.get(page).text,'lxml')
    airlines_name = airlines_soup.select('.title h1')
    airlines = airlines_name[0].getText()
    airlines_fleet = pd.read_html(page)
    seat_config = pd.concat(airlines_fleet)
    seat_config['airlines'] = airlines
    seat_df = seat_df.append(seat_config) #Gộp các df thành 1 df chính, phải dùng df = df.append() để store output do method append ko xảy ra in-place
    # Tìm các link fleetinfo để scrape các bảng chất lượng ghế
    fleet_info_page = airlines_soup.find_all('a', href = re.compile('fleetinfo'))
    fleet_info_pages.append('https://seatguru.com'+fleet_info_page[0]['href'])
seat_df.to_excel(r'C:\Users\namtrantuan\Desktop\Python\seat_config.xlsx')
```

```python
fleet_info_pages
```

```python
seat_quality_df = pd.DataFrame()
for page in fleet_info_pages:
    fleet_info = bs4.BeautifulSoup(requests.get(page).text,'lxml')
    khoangdv = fleet_info.select('.class-of-service')
    acinfo = pd.read_html(page)
    service_class = []
    for khoang in khoangdv:
        service_class.append(khoang.getText())
    for table in acinfo:
        table['khoangdv'] = service_class.pop(0)
        table['airline'] = page
    seat_quality = pd.concat(acinfo)
    seat_quality_df = seat_quality_df.append(seat_quality)
seat_quality_df.to_excel(r'C:\Users\namtrantuan\Desktop\Python\seat_quality.xlsx')
```
